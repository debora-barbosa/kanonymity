```python
from faker import Faker
import pandas as pd
import random
import numpy as np

fake = Faker("pt_BR")

# quantidade de cadastros
n = 1000
alunos = list()

for i in range(n):
    alunos.append([
        fake.first_name() + " " + fake.last_name(), 
        random.randint(14,19), 
        fake.address(),
        fake.random_number(digits=9, fix_len=True), 
        random.choice(["1º", "2º", "3º"]), 
        random.choice(["A", "B"]), 
        round(random.uniform(0,10),2), 
        round(random.uniform(0,10),2), 
        round(random.uniform(0,10),2), 
        round(random.uniform(0,10),2), 
        round(random.uniform(0,10),2)
    ])

df = pd.DataFrame(alunos, columns=[
    'Nome','Idade', 'Endereço', 'Matrícula','Ano Escolar','Turma',
    'Média em Ciências da Natureza',
    'Média em Ciências Humanas',
    'Média em Matemática',
    'Média em Linguagens',
    'Média em Redação'
])

print(df.head())

#deletar os identificadores do dataframe
df2 = df.copy()
df2 = df2.drop(['Nome', 'Matrícula', 'Endereço'], axis=1)

#anonimizar quase identificadores

df2['Idade'] = pd.cut(df2['Idade'],bins=[13,16,19],labels=['14-16','17-19'])
df2['Média em Ciências da Natureza'] = pd.cut(df2['Média em Ciências da Natureza'],bins=[0,5,10.1],labels=['0-5','5-10'])
df2['Média em Ciências Humanas'] = pd.cut(df2['Média em Ciências Humanas'],bins=[0,5,10.1],labels=['0-5','5-10'])
df2['Média em Matemática'] = pd.cut(df2['Média em Matemática'],bins=[0,5,10.1],labels=['0-5','5-10'])
df2['Média em Linguagens'] = pd.cut(df2['Média em Linguagens'],bins=[0,5,10.1],labels=['0-5','5-10'])
df2['Média em Redação'] = pd.cut(df2['Média em Redação'],bins=[0,5,10.1],labels=['0-5','5-10'])
df2['Ano Escolar'] = 'Ensino Médio'
df2['Turma'] = '*'
print(df2.head())

k = df2.groupby(df2.columns.tolist(), observed=True).size().min()
print("Valor mínimo de k:", k)
```
